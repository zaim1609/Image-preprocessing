```python

import matplotlib.pyplot as plt
from PIL import Image,ImageEnhance

img = Image.open("../input/dataset-tutorial-pcd/pollen_low_contrast.jpg")
lbr,tgi = (img.size)


cg = Image.new('L',(lbr,tgi),(0))

alp = 1.5
for x in range(lbr):
    for y in range(tgi):
        fxy = img.getpixel((x,y))
        gxy = int(alp*fxy)
        cg.putpixel((x,y),(gxy))
plt.imshow(cg, cmap='gray')
```
